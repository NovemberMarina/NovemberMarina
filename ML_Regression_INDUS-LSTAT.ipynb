```python
X = df[["INDUS"]].values
Y = df[["LSTAT"]].values

# 【前処理】標準化
# 回帰問題は、変数を正規化が予測精度を寄与している
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()

X_std = sc_x.fit_transform(X)
Y_std = sc_y.fit_transform(Y).flatten()
```

```python
"""
前提：コスト関数（誤差関数）を最小化するように重みを更新する
結果：誤差が最も少ない線形関数を求めること
"""
class LinearRegressionGD:
    
    def __init__(self, eta=0.0001, n_iter=10):
        self.eta = eta        # 学習率
        self.n_iter = n_iter  # 学習回数
        
        # ====================================
        self.index = []
        self.X_s = []
        self.Y_s = []
        self.ouput_s  = []
        self.errors_s = []
        self.w0_s = []
        self.w1_s = []
        self.cost_s = []
        # ====================================
        
    def fit(self, X, Y):
        self.w_ = np.zeros(1 + X.shape[1]) # 重みの宣言
        self.cost_ = []                    # 誤差関数の宣言
        
        
        for i in range(self.n_iter):
            ouput = self.net_input(X)  # 線形関数
            errors = (Y - ouput)       # 誤差
            
            self.w_[1:] += self.eta * X.T.dot(errors) # W1 重み べクトルの内積（w1*x）
            self.w_[0:] += self.eta * errors.sum()    # W0 重み ベクトルの内積（w0*1）
            cost = (errors**2).sum() / 2.0            # 誤差平方差：（実績値ー予測値）**2 / 2
            self.cost_.append(cost)                   # 誤差関数を記録
            
            # ====================================
            self.index.append(i+1)
            self.X_s.append(X[0])
            self.Y_s.append(Y[0])
            self.ouput_s.append(ouput[0])
            self.errors_s.append(errors[0])
            self.w0_s.append(self.w_[1:][0])
            self.w1_s.append(self.w_[1:][0])
            self.cost_s.append(cost)
            # ====================================
        
        # ====================================
        calc_df = pd.DataFrame()
        calc_df["No"] = self.index
        calc_df["X"] = self.X_s
        calc_df["Y"] = self.Y_s
        calc_df["ouput"] = self.ouput_s
        calc_df["errors"] = self.errors_s
        calc_df["w0"] = self.w0_s
        calc_df["w1"] = self.w1_s
        calc_df["cost"] = self.cost_s
        display(calc_df)
        # ====================================
        
        return self
    
    def net_input(self, X):
        return self.w_[0] + np.dot(X, self.w_[1:])
    
    def predict(self, X):
        return self.net_input(X)
    

lr = LinearRegressionGD()
lr.fit(X_std, Y_std)



def liner_func(x):
    print("求めた線形関数")
    print("y = {0} + {1}x".format(lr.w_[0], lr.w_[1]))
    y = lr.w_[0] + lr.w_[1]*x
    return y

liner_func(1.2)
```

```python
plt.plot(range(1, lr.n_iter+1), lr.cost_)
plt.ylabel("SSE")
plt.xlabel("Epoch")
plt.show()
```

```python
def lin_regplot(X, Y, model):
    plt.scatter(X, Y, c="steelblue", edgecolor="white", s=70)
    plt.plot(X, model.predict(X), color="black", lw=2)
    plt.xlabel('Rate of non-retail business')
    plt.ylabel('Rate of low-income')
    plt.show()
    return 

lin_regplot(X_std, Y_std, lr)
```
